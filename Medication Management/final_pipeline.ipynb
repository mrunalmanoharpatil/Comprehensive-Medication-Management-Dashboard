```python
# Step 1: Create a copy to work safely
df_allergies_cleaned = df_A_imp.copy()

# Step 2: Aggregate
df_allergy_agg = df_allergies_cleaned.groupby('PATIENT').agg({
    'START': 'min',  # earliest allergy record
    'DESCRIPTION': lambda x: ', '.join(x.dropna().unique()),
    'DESCRIPTION1': lambda x: ', '.join(x.dropna().unique()),
    'DESCRIPTION2': lambda x: ', '.join(x.dropna().unique()),
    'SEVERITY1': lambda x: ', '.join(x.dropna().unique()),
    'SEVERITY2': lambda x: ', '.join(x.dropna().unique())
}).reset_index()

# Optional: Rename for clarity
df_allergy_agg.rename(columns={
    'START': 'ALLERGY_START_DATE',
    'DESCRIPTION': 'ALLERGIES',
    'DESCRIPTION1': 'REACTION_DESC_1',
    'DESCRIPTION2': 'REACTION_DESC_2',
    'SEVERITY1': 'SEVERITY_1',
    'SEVERITY2': 'SEVERITY_2'
}, inplace=True)

df_allergy_agg.info()
```

```python
df_med_agg = df_M_imp.groupby("PATIENT").agg({
    'START': 'min',
    'STOP': 'max',
    'DESCRIPTION': lambda x: ', '.join(x.dropna().unique()),
    'TOTALCOST': 'sum',
    'DISPENSES': 'sum',
    'REASONDESCRIPTION': lambda x: ', '.join(x.dropna().unique()),
    'DAYS_SUPPLY': 'sum',
    'allergy_conflict': 'any'
}).reset_index()

# Optional rename
df_med_agg.rename(columns={
    'START': 'MED_START_DATE',
    'STOP': 'MED_STOP_DATE',
    'DESCRIPTION': 'MEDICATIONS',
    'REASONDESCRIPTION': 'MEDICATION_REASONS',
    'DAYS_SUPPLY': 'TOTAL_DAYS_SUPPLIED',
    'allergy_conflict': 'HAS_ALLERGY_CONFLICT'
}, inplace=True)

df_med_agg.info()
```

```python
df_C_imp.info()
```

```python
df_cond_agg = df_C_imp.groupby("PATIENT").agg({
    'START': 'min',
    'STOP': 'max',
    'DESCRIPTION': lambda x: ', '.join(x.dropna().unique()),
    'CODE': lambda x: ', '.join(str(code) for code in x.dropna().unique()),
}).reset_index()

# Optional: count of conditions
df_cond_agg['NUM_CONDITIONS'] = df_C_imp.groupby('PATIENT')['DESCRIPTION'].count().values

# Optional rename
df_cond_agg.rename(columns={
    'START': 'COND_FIRST_DATE',
    'STOP': 'COND_LAST_DATE',
    'DESCRIPTION': 'CONDITIONS',
    'CODE': 'CONDITION_CODES'
}, inplace=True)
df_cond_agg.info()
```

```python
df_merged = df_P_imp.merge(df_allergy_agg, how='left', left_on='Id', right_on='PATIENT')

# Merge with medication aggregates
df_merged = df_merged.merge(df_med_agg, how='left', left_on='Id', right_on='PATIENT', suffixes=('_allergy', '_med'))

# Merge with condition aggregates
df_merged = df_merged.merge(df_cond_agg, how='left', left_on='Id', right_on='PATIENT')

# Drop extra PATIENT columns to avoid confusion (since they are all the same after merging)
df_merged = df_merged.drop(columns=['PATIENT_allergy', 'PATIENT_med', 'PATIENT'])

# Optional: rename the main PATIENT ID column if needed
df_merged = df_merged.rename(columns={'Id': 'PATIENT_ID'})

# View the final merged dataframe
df_merged.head()
```

```python
import pandas as pd

def clean_datetime_for_mongo(df):
    for col in df.columns:
        if pd.api.types.is_datetime64_any_dtype(df[col]):
            # Convert timezone-aware to naive
            if df[col].dt.tz:
                df[col] = df[col].dt.tz_localize(None)
            # Replace NaT with None
            df[col] = df[col].apply(lambda x: x if pd.notnull(x) else None)
    return df
```

```python
df_cleaned = clean_datetime_for_mongo(df_merged)
df_cleaned['BIRTHDATE'] = df_cleaned['BIRTHDATE'].astype(str)
df_cleaned['ALLERGY_START_DATE'] = df_cleaned['ALLERGY_START_DATE'].astype(str)
df_cleaned['MED_START_DATE'] = df_cleaned['MED_START_DATE'].astype(str)
df_cleaned['MED_STOP_DATE'] = df_cleaned['MED_STOP_DATE'].astype(str)
df_cleaned['COND_FIRST_DATE'] = df_cleaned['COND_FIRST_DATE'].astype(str)
df_cleaned['COND_LAST_DATE'] = df_cleaned['COND_LAST_DATE'].astype(str)

```
